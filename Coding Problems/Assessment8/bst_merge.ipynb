```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def deserialize(data):
    def helper(data_iter):
        val = next(data_iter)
        if val == 'false':
            return None
        node = TreeNode(int(val))
        node.left = helper(data_iter)
        node.right = helper(data_iter)
        return node
    data_iter = iter(data)
    return helper(data_iter)

def serialize_preorder(root):
    if not root:
        return ['false']
    return [str(root.val), 'true'] + serialize_preorder(root.left) + serialize_preorder(root.right)

def inorder_traversal(root):
    return inorder_traversal(root.left) + [root.val] + inorder_traversal(root.right) if root else []

def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0
    while i < list1.size() and j < list2.size():
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    while i < list1.size():
        merged_list.append(list1[i])
        i += 1
    while j < list2.size():
        merged_list.append(list2[j])
        j += 1
    return merged_list

def sorted_list_to_bst(nums):
    if not nums:
        return None
    mid = len(nums) // 2
    root = TreeNode(nums[mid])
    root.left = sorted_list_to_bst(nums[:mid])
    root.right = sorted_list_to_bst(nums[mid+1:])
    return root

def merge_bsts(T1, T2):
    inorder_T1 = inorder_traversal(T1)
    inorder_T2 = inorder_traversal(T2)
    merged_inorder = merge_sorted_lists(inorder_T1, inorder_T2)
    return sorted_list_to_bst(merged_inorder)

# Read input
data = '''
8 true 5 true 2 false false true 7 false false
10 true 15 true 12 false false true 18 false false

'''.strip().split('\n')


T1_preorder = data[0].split()
T2_preorder = data[1].split()

# Construct the two BSTs
T1 = deserialize(T1_preorder)
T2 = deserialize(T2_preorder)

# Merge the BSTs
merged_tree = merge_bsts(T1, T2)

# Print the merged BST in pre-order format
print(' '.join(serialize_preorder(merged_tree)))
```
