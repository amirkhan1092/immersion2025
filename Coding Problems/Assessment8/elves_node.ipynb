```python
class Node:
    def __init__(self, data):
        self.data = data
        self.right = None
        self.down = None

def merge(a, b):
    if not a:
        return b
    if not b:
        return a
    
    if a.data < b.data:
        result = a
        result.down = merge(a.down, b)
    else:
        result = b
        result.down = merge(a, b.down)
    
    return result

def flatten(root):
    if not root or not root.right:
        return root
    
    root.right = flatten(root.right)
    root = merge(root, root.right)
    
    return root

def print_flattened_list(head):
    while head:
        print(head.data, end=" ")
        head = head.down
    print()

def create_linked_list(data, heads):
    if not data:
        return None
    
    head = Node(data.pop(0))
    heads.append(head)
    
    current = head
    while data:
        current.down = Node(data.pop(0))
        current = current.down
    
    return head

if 1:
    data1 = '''

2
2 3                  
10 20 19 22 50
'''.strip().split('\n')
    
    N = int(data1[0])

    
    if N == 0:
        print()
        
    
    head_counts = list(map(int, data1[1].split()))
    
    data = list(map(int, data1[2].split()))
    
    heads = []
    create_linked_list(data[:head_counts[0]], heads)
    start = heads[0]
    
    for i in range(1, N):
        create_linked_list(data[sum(head_counts[:i]):sum(head_counts[:i+1])], heads)
    
    for i in range(N-1):
        heads[i].right = heads[i+1]
    
    start = flatten(start)
    print_flattened_list(start)


```
