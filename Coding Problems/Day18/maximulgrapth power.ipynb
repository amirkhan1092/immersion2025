```python
def maximal_power(n, edges):
    from collections import defaultdict

    # Create adjacency list
    graph = defaultdict(set)
    for edge in edges:
        a, b = edge
        graph[a].add(b)
        graph[b].add(a)
    
    max_power = 0
    
    # Iterate over all pairs of vertices
    for i in range(n):
        for j in range(i + 1, n):
            # Combine the edges connected to i and j
            combined_edges = graph[i] | graph[j]
            # Count the number of unique edges
            power = len(combined_edges)
            # Update max power if necessary
            max_power = max(max_power, power)
    
    return max_power

# Input reading
data = '''

12 54
7 0
6 0
7 9
7 5
2 1
6 3
8 1
4 6
6 4
8 0
7 10
8 7
4 5
3 10
10 3
8 9
5 6
3 7
4 1
10 8
8 3
1 8
4 6
2 4
1 8
4 7
3 2
10 6
11 12
10 7
2 9
9 1
7 8
3 11
1 0
3 4
5 2
11 1
11 0
4 5
9 10
7 11
10 8
7 6
9 6
9 2
11 1
5 9
7 6
5 8
3 5
5 8
7 11
5 2


'''.strip().split('\n')
n, m = map(int, data[0].split())
edges = [tuple(map(int, data[_].split())) for _ in range(1, m+1)]

# Calculate and print the maximal power
print(maximal_power(n, edges))
```
