```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

def build_tree(preorder):
    def helper(it):
        try:
            val = next(it)
        except StopIteration:
            return None
        if val == 'false':
            return None
        node = TreeNode(int(val))
        if next(it) == 'true':
            node.left = helper(it)
        else:
            next(it)  # Skip 'false'
            node.left = None
        if next(it) == 'true':
            node.right = helper(it)
        else:
            next(it)  # Skip 'false'
            node.right = None
        return node

    return helper(iter(preorder.split()))

def inorder_traversal(root, nodes):
    if root is not None:
        inorder_traversal(root.left, nodes)
        nodes.append(root)
        inorder_traversal(root.right, nodes)

def recover_bst(preorder):
    root = build_tree(preorder)
    
    # Step 1: Perform inorder traversal to find the swapped nodes
    nodes = []
    inorder_traversal(root, nodes)
    
    first = None
    second = None
    for i in range(len(nodes) - 1):
        if nodes[i].val > nodes[i + 1].val:
            if first is None:
                first = nodes[i]
            second = nodes[i + 1]
    
    # Step 2: Swap the values of the two nodes
    if first and second:
        first.val, second.val = second.val, first.val
    
    # Step 3: Print the corrected tree in preorder
    print_preorder(root)
    print()

def print_preorder(root):
    if root is None:
        print("false", end=" ")
    else:
        print(root.val, end=" ")
        if root.left is not None or root.right is not None:
            print("true", end=" ")
            print_preorder(root.left)
            print_preorder(root.right)
        else:
            print("false false", end=" ")

# Sample Inputs
preorder1 = "8 true 6 true 5 false false true 7 false false true 10 false true 12 false false"
recover_bst(preorder1)

preorder2 = "12 true 8 true 5 false false true 10 true 9 false false true 11 false false true 15 false true 14 false false true 17 false false"
recover_bst(preorder2)

# Extra Test Cases
preorder3 = "25 true 20 true 15 false false true 22 false false true 30 true 28 false false true 35 false false"
recover_bst(preorder3)

preorder4 = "18 true 15 true 10 true 5 false false true 12 false false true 17 true 16 false false true 20 true 19 false false true 25 false false"
recover_bst(preorder4)

preorder5 = "50 true 40 true 35 false false true 45 false false true 60 true 55 false false true 65 false false"
recover_bst(preorder5)
```
